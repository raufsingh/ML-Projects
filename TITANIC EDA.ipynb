# operations


# 1.handling missing values  


```python
#1.drop null values
titanic.drop('Cabin',axis=1,inplace=True)
```

```python
titanic.isnull().sum()
```

```python
#2.fill null values
#.fillna()
```

```python
#titanic['Age'].mode()
titanic['Age'].mean()
#titanic['Age'].median()
```

```python
titanic["Age"].fillna(29,inplace=True)#method='ffill'
```

```python
titanic.isnull().sum()
```

```python
titanic['Embarked'].mode()
```

```python
titanic["Embarked"].fillna("S",inplace=True)
```

```python
titanic.isnull().sum()
```
